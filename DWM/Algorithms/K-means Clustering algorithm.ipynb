```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Load the Iris dataset
iris = load_iris()
X = iris.data

# 2. Choose number of clusters (for Iris, set n_clusters=3)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# 3. Get the cluster labels for each sample
labels = kmeans.labels_
print("Cluster labels for each record: ", labels)

# 4. (Optional) Compare clusters with real classes
print("Actual iris classes: ", iris.target)

# 5. Visualize clusters (using first two features)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("K-means clustering (Iris data)")
plt.show()

# 6. See cluster centers
print("Cluster Centers:\n", kmeans.cluster_centers_)
```

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Step 1: Generate sample data
X, y = make_blobs(n_samples=200, centers=3, cluster_std=0.5, random_state=0)

# Step 2: Visualize the original data points
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Original Data Points")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Step 3: Apply K-Means algorithm
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Step 4: Get cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Step 5: Visualize clustered data
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, marker='X')
plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Step 6: Display results
print("Cluster Centers:\n", centers)
print("\nLabels Assigned to Each Data Point:\n", labels)
```
